```python
import pandas as pd

# Load the source data
source_data = pd.read_csv(r"C:\Users\gagan\Downloads\Data\Winter 23-24\Capstone\fraud-transactions-detection\JPMorgan\SupervisedFraudData.csv")

# Typecasting columns
# source_data['Time_step'] = source_data['Time_step'].astype(str)
source_data['Transaction_Id'] = source_data['Transaction_Id'].astype(str)
source_data['Sender_Id'] = source_data['Sender_Id'].astype(str)
source_data['Sender_Account'] = source_data['Sender_Account'].astype(str)
source_data['Sender_Country'] = source_data['Sender_Country'].astype(str)
source_data['Sender_Sector'] = source_data['Sender_Sector'].astype(str)
source_data['Sender_lob'] = source_data['Sender_lob'].astype(str)
source_data['Bene_Id'] = source_data['Bene_Id'].astype(str)
source_data['Bene_Account'] = source_data['Bene_Account'].astype(str)
source_data['Bene_Country'] = source_data['Bene_Country'].astype(str)
source_data['USD_amount'] = source_data['USD_amount'].astype(str)
source_data['Label'] = source_data['Label'].astype(str)
source_data['Transaction_Type'] = source_data['Transaction_Type'].astype(str)

# Deduplicate data
deduplicated_data = source_data.drop_duplicates()

# Fill null values
filled_data = deduplicated_data.fillna({
    'Sender_Id': 'Unknown',
    'Sender_Account': 'Unknown',
    'Sender_Country': 'Unknown',
    'Sender_Sector': 0,
    'Sender_lob': 'Unknown',
    'Bene_Id': 'Unknown',
    'Bene_Account': 'Unknown',
    'Bene_Country': 'Unknown'
})

# Split Time_step into Date and Time
# filled_data['Date'] = source_data['Time_step'].str.split(" ").str[0]
# filled_data['Time'] = source_data['Time_step'].str.split(" ").str[1]

# Typecast columns
filled_data['Sender_Sector'] = filled_data['Sender_Sector'].astype(int)
filled_data['Label'] = filled_data['Label'].astype(int)
filled_data['USD_amount'] = filled_data['USD_amount'].astype(float)
# filled_data['Date'] = pd.to_datetime(filled_data['Date'])
# filled_data['Time'] = pd.to_timedelta(filled_data['Time'].apply(lambda x: int(x.split(":")[1]) * 3600 + int(x.split(":")[2]) * 60 + int(x.split(":")[3])))

# Split the data into two batches based on Label
# supervised_data = filled_data.drop(['Transaction_Id', 'Sender_Id', 'Sender_Account', 'Sender_Country', 'Sender_Sector', 'Sender_lob', 'Bene_Id', 'Bene_Account', 'Bene_Country', 'USD_amount', 'Transaction_Type'], axis=1)
unsupervised_data = filled_data

# Save the processed data to CSV files
# supervised_data.to_csv("SupervisedFraudData.csv", index=False)
unsupervised_data.to_csv("UnsupervisedFraudData.csv", index=False)
```
